# Distancias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.spatial import distance_matrix
```

```python
df = pd.read_csv("python-ml-course-master/datasets/movies/movies.csv",sep=";")
```

```python
df
```

```python
#todo esto solo para eliminar el user_id
cols = df.columns.values.tolist()[1:]
cols
```

```python
#Distancia de manhatand
dd1 = distance_matrix(df[cols],df[cols],p=1)
#Distancia euclidia
dd2 = distance_matrix(df[cols],df[cols],p=2)
#Distancia X (inventada)
dd10 = distance_matrix(df[cols],df[cols],p=10)
```

```python
dd1
```

```python
def dm_to_df(dd,col_name):
    return pd.DataFrame(dd,index=col_name,columns=col_name)
```

```python
dm_to_df(dd1,df["user_id"])
```

```python
dm_to_df(dd2,df["user_id"])
```

```python
dm_to_df(dd10,df["user_id"])
```

# Clustering jerrquico en Python

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df[cols]
```

```python
Z = linkage(df[cols], "ward")
Z
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerarquico para el clustering")
plt.xlabel("ID de los usuarios")
plt.ylabel("Distancia")
dendrogram(Z,leaf_rotation=90.,leaf_font_size=20)
plt.show()
```

```python
Z = linkage(df[cols], "average")
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerarquico para el clustering")
plt.xlabel("ID de los usuarios")
plt.ylabel("Distancia")
dendrogram(Z,leaf_rotation=90.,leaf_font_size=20)
plt.show()
```

```python
Z = linkage(df[cols], "complete")
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerarquico para el clustering")
plt.xlabel("ID de los usuarios")
plt.ylabel("Distancia")
dendrogram(Z,leaf_rotation=90.,leaf_font_size=20)
plt.show()
```

```python
 Z = linkage(df[cols], "single")
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerarquico para el clustering")
plt.xlabel("ID de los usuarios")
plt.ylabel("Distancia")
dendrogram(Z,leaf_rotation=90.,leaf_font_size=20)
plt.show()
```
