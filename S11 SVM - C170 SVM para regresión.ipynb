# SVM para regresi√≥n

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.sort(5*np.random.rand(200,1),axis=0)
Y = np.sin(X).ravel()
```

```python
Y[::5] += 3*(0.5 - np.random.rand(40))
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X,Y, color="darkorange",label="data")
```

```python
from sklearn.svm import SVR
```

```python
C=1e3
svr_lin = SVR(kernel="linear", C=C)
svr_rbf = SVR(kernel="rbf", C=C, gamma=0.1)
svr_pol = SVR(kernel="poly", C=C, degree=3)
```

```python
y_lin = svr_lin.fit(X,Y).predict(X)
y_rbf = svr_rbf.fit(X,Y).predict(X)
y_pol = svr_pol.fit(X,Y).predict(X)
```

```python
lw = 2
plt.figure(figsize=(16,9))
plt.scatter(X, Y, color="darkorange", label="data")
plt.plot(X, y_lin, color="navy", lw=lw, label="linear")
plt.plot(X, y_rbf, color="c", lw=lw, label="radial")
plt.plot(X, y_pol, color="y", lw=lw, label="poly")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("SVR")
plt.legend()
```
