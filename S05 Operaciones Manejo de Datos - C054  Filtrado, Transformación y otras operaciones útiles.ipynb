```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gender = ['Male', 'Female']
income = ['Poor', 'Middle Class', 'Rich']
```

```python
n = 500

gender_data = []
income_data = []

for i in range(0,500):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
    
height = 160 + 30 * np.random.randn(n)
weigth = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
gender_data[1:10]
```

```python
income_data[1:10]
```

```python
df = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weigth" : weigth,
        "Age" : age,
        "Income" : income
    }
)
```

```python
df.head()
```

## Agrupación de datos

```python
grouped_gender = df.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
```

```python
double_group = df.groupby(["Gender","Economic Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

## Operaciones de agregacion

```python
double_group.sum()
```

```python
double_group.mean()
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
double_group["Income"].describe()
```

```python
double_group.aggregate(
    {
        "Income" : np.sum,
        "Age" : np.mean,
        "Height" : np.std
    }
)
```

```python
double_group.aggregate(
    {
        "Age" : np.mean,
        "Height" : lambda h:(np.mean(h)) / np.std(h)
    }
)
```

```python
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
double_group.aggregate([lambda x: np.median(x) / np.std(x)])
```

## Filtrado

```python
double_group["Age"].filter(lambda x: x.sum() > 2400)
```

## Transformacion

```python
zscore = lambda x : (x - x.mean()) - x.std()
```

```python
double_group.transform(zscore)
```

```python
z_group = double_group.transform(zscore)
```

```python
plt.hist(z_group["Age"])
```

## funciona de completado de datos.. por ejemplo con el promedio

```python
fill_na_mean = lambda x: x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean)
```
