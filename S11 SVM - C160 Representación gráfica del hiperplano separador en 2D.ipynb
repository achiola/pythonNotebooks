# Soporte Vectorial Clasificador Linea

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn import svm
```

```python
X = [1, 5, 1.5, 8, 1, 9]
Y = [2, 8, 1.8, 8, 0.6, 11]
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X,Y)
plt.show()
```

# Creando el modelo clasificador lineal

```python
data = np.array(list(zip(X,Y)))
```

```python
data
```

## Etiquetamos

```python
target = [0, 1, 0, 1, 0, 1]
```

```python
classifier = svm.SVC(kernel="linear", C=1.0)
classifier.fit(data, target)
```

```python
classifier.predict(np.array([0.57, 0.67]).reshape(1,2))
```

```python
classifier.predict(np.array([10, 11]).reshape(1,2))
```

# Representación gráfica del hiperplano separador en 2D

```python
# Obtenemos los pesos
w = classifier.coef_[0]
w
```

```python
a = -w[0] / w[1]
a
```

```python
b = -classifier.intercept_[0] / w[1]
b
```

## y = a * x + b

```python
xx = np.linspace(0,10)
yy = a * xx + b
```

```python
plt.figure(figsize=(10,8))
plt.plot(xx,yy,'k-',label="Hiperplano de separacion")
plt.scatter(X,Y, c=target)
plt.legend()
plt.show()
```
